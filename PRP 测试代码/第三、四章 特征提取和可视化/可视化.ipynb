```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
```

```python
# 坐标轴设置
# 创建图像对象和坐标轴
fig = plt.figure(figsize = (10,6))
plt.axis([-1,1,-10,10]) #xmin, xmax, ymin, ymax
# 调整坐标的刻度
# 1.编辑locator和formatter
ax = plt.subplot()
xmajorLocator = MultipleLocator(2.5) #x轴刻度设置为2.5的倍数
xmajorFormatter = FormatStrFormatter('%.1f') #数字格式为保留一位小数
ymajorLocator = MultipleLocator(20)
ymajorFormatter = FormatStrFormatter('%.1f')
# 2.设置图表刻度
ax.xaxis.set_major_locator(xmajorLocator)
ax.xaxis.set_major_formatter(xmajorFormatter)
ax.yaxis.set_major_locator(ymajorLocator)
ax.yaxis.set_major_formatter(ymajorFormatter)
ax.set_xlim(left = 0, right = 10)
ax.set_ylim(bottom = 0)

#作一个抛物线图
x = range(10)
y = [i**2 for i in x]
plt.plot(x,y)
plt.autoscale() #用于自动调整y轴的数据范围（我们没有手动设置top）
```

```python
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False

fig = plt.figure(figsize = (12,8))
ax1 = plt.subplot()
x = np.arange(1e-3, np.e, 0.001)
y1 = np.exp(-x)
y2 = np.log(x)

ax1.plot(x,y1)
ax1.set_ylabel('主轴')
ax1.set_xlabel('横轴')

ax2 = ax1.twinx() #twinx()会在图表的镜像边生成一个新轴
ax2.plot(x,y2)
ax2.set_xlim([0, np.e])
ax2.set_ylabel('次轴')
```
