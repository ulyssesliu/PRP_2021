# 1. 统计量分析
## 分为**集中趋势度量**和**离中趋势度量**

```python
#速度分析
data_speed = data[data['time_id']==1]['aveSpeed']
speed_statistics = data_speed.describe() #保存基本统计量
speed_statistics.loc['range'] = speed_statistics.loc['max'] - speed_statistics.loc['min'] #极差
speed_statistics.loc['cv'] = speed_statistics.loc['std']/speed_statistics.loc['mean'] #变异系数
speed_statistics.loc['dis'] = speed_statistics.loc['75%'] - speed_statistics.loc['25%'] #四分位数间隔
```

# 2.对比分析
### a.绝对数比较：直接比绝对值
### b.相对数比较：用计算出的相对数表征联系
- 结构相对数：占整体的比重
- 比较相对数：同一时间、不同空间，比如不同地区的商品价格，不同企业的某项指标
- 强度相对数：就是计算密度，比如公交线路密度km/km^2
- 计划完成程度相对数：完成数与计划数的比值
- 动态相对数：同一现象在不同时期的比值，比如发展速度、增长速度
